# 1- Print a Square of stars

```python
def printsquarestars(rows):
    for i in range(rows):
        print('*'*rows)
rows=int(input("Enter the Number of rows:"))
printsquarestars(rows)
```

# 2- Print a Right Triangle Pattern of Stars

```python
def triangle(rows):
    for i in range(1,rows+1):
        for j in range(0,i):
            print('*', end='')
        print()
n = int(input())
triangle(n)
```

# 3- Print a Hollow Square Pattern of stars


```python

def hollow(rows,cols):
    for i in range(rows):
        for j in range(cols):
            if i == 0 or i == rows - 1 or j == 0 or j == cols - 1:
                print("*", end=" ")
            else:
                print(" ", end=" ")
        print()

r = int(input("Rows: "))
c = int(input("Columns: "))
hollow(r,c)
```

# 4- Print a Staircase Pattern of stars

```python
def stairs(n):
    for i in range(1,n+1):
        for j in range(2):
            for k in range(i*2):
                print('*', end='')
            print()
n = int(input())
stairs(n)
```

# 5- Print a Butterfly Pattern of Stars

```python
def butterfly(n):
    # Upper part
    for i in range(1, n + 1):
        for j in range(1, n * 2):
            if j <= i or (n*2)-j <= i:
                print("*", end="")
            else:
                print(" ", end="")
        print()

    # Lower part
    for i in range(n-1, 0, -1):
        for j in range(1, n * 2):
            if j <= i or (n*2)-j <= i:
                print("*", end="")
            else:
                print(" ", end="")
        print()

butterfly(5)
```
