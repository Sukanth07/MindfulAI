1 - Given an integer array nums, return true if any value appears at least twice in the array, and return false if every element is distinct.

 

```python
# Example 1:

# Input: nums = [1,2,3,1]
# Output: true
# Example 2:

# Input: nums = [1,2,3,4]
# Output: false
# Example 3:

# Input: nums = [1,1,1,3,3,4,3,2,4,2]
# Output: true

def distinct_elements(nums):
    hmap = {}
    for i in range(len(nums)):
        if nums[i] not in hmap:
            hmap[nums[i]] = 1
        elif nums[i] in hmap:
            hmap[nums[i]] += 1
        if hmap[nums[i]] == 2:
            return True
    return False

nums = [1,2,3,1]
print(distinct_elements(nums))
```

2 - Given two strings s and t, return true if t is an anagram of s, and false otherwise.

An Anagram is a word or phrase formed by rearranging the letters of a different word or phrase, typically using all the original letters exactly once.

 

```python
# Example 1:

# Input: s = "anagram", t = "nagaram"
# Output: true
# Example 2:

# Input: s = "rat", t = "car"
# Output: false

def isAnagram(s, t):
    hmap1 = {}
    hmap2 = {}
    for i in s:
        if i not in hmap1: hmap1[i] = 1
        elif i in hmap1: hmap1[i] += 1
    for i in t:
        if i not in hmap2: hmap2[i] = 1
        elif i in hmap2: hmap2[i] += 1
    if hmap1 == hmap2:
        return True
    return False    

s = 'anagram'
t = 'nagaram'
print(isAnagram(s,t))
```

3 - Given a List. The task is to find the sum and average of the list. The average of the list is defined as the sum of the elements divided by the number of elements.

```python
# Input: [4, 5, 1, 2, 9, 7, 10, 8]
# Output:
# sum =  46
# average =  5.75

# Input: [15, 9, 55, 41, 35, 20, 62, 49]
# Output:
# sum =  286
# average =  35.75

def calculate(nums):
    n = len(nums)
    total = 0
    for i in nums:
        total += i
    avg = total / n
    return total, avg

nums = [4, 5, 1, 2, 9, 7, 10, 8]
sum_ , avg_ = calculate(nums)
print('Sum: ', sum_)
print('Avg: ', avg_)
```

5- Given an array of N strings, find the longest common prefix among all strings present in the array.

```python
# Input:
# N = 4
# arr[] = {geeksforgeeks, geeks, geek,
#          geezer}
# Output: gee
# Explanation: "gee" is the longest common
# prefix in all the given strings.

def longest(arr):
    res = ""
    str_0_len = len(arr[0])
    for i in range(str_0_len):
        for s in arr:
            if i==len(s) or arr[0][i] != s[i]:
                return res
        res += arr[0][i]

def longest_prefix(arr):
    min_len = len(min(arr, key=len))
    min_len_index = arr.index(min(arr, key=len))
    flag = False
    for i in range(min_len):
        for j in range(len(arr)):
            if arr[min_len_index][:min_len] == arr[j][:min_len]:
                flag = True
            else:
                flag = False
                min_len -= 1
                break
        if flag == True:
            return arr[min_len_index][:min_len]


arr = ['geeksforgeeks', 'geeks', 'geek', 'geezer']
print(longest(arr))
```

6- Given an array a of size N which contains elements from 0 to N-1, you need to find all the elements occurring more than once in the given array. Return the answer in ascending order. If no such element is found, return list containing [-1]. 

Note: The extra space is only for the array to be returned. Try and perform all operations within the provided array. 

```python
# Input:
# N = 4
# a[] = {0,3,1,2}
# Output: 
# -1
# Explanation: 
# There is no repeating element in the array. Therefore output is -1.
from collections import deque
def occurance(arr):
    hmap = {}
    for i in range(len(arr)):
        if arr[i] not in hmap:
            hmap[arr[i]] = 1
        elif arr[i] in hmap:
            hmap[arr[i]] += 1
    arr = []
    for key,value in hmap.items():
        if value > 1:
            arr.append(key)
    if len(arr) == 0:
        return -1
    return sorted(arr)

arr = [0, 3, 1, 2, 0, 3, 2]
print(occurance(arr))
```
