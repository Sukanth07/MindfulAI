```python
# Singly Linked List

class Node:
    def __init__(self, data) -> None:
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self) -> None:
        self.head = None

    def insertAtBeginning(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            new_node.next = self.head
            self.head = new_node
        
        self.displayLinkedList()

    def insertAtEnd(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            temp = self.head
            while temp.next is not None:
                temp = temp.next
            temp.next = new_node
        
        self.displayLinkedList()


    def insertAtIndex(self, data, index):
        if index == 0:
            self.insertAtBeginning(data)
            return
        
        if self.head == None:
            return "Head Node is not initialized!"
        
        new_node = Node(data)
        pos = 0
        temp = self.head
        while temp != None and pos+1 != index:
            pos += 1
            temp = temp.next

        if temp != None:
            new_node.next = temp.next
            temp.next = new_node
        else:
            print(f"Index {index} is not valid!")
            return
        
        self.displayLinkedList()


    def insertAfterElement(self, data, element):
        if self.head == None:
            print("Head Node is not initialized!")
            return
        new_node = Node(data)
        temp = self.head
        while temp != None:
            if temp.data != element:
                temp = temp.next
            else:
                break
        if temp:
            new_node.next = temp.next
            temp.next = new_node
        else:
            print(f"Element \'{element}\' is not found!")
            return
        
        self.displayLinkedList()
            


    def deleteAtBeginning(self):
        if self.head == None:
            print("Head Node is not initialized!!")
            return 
        
        self.head = self.head.next
        
        self.displayLinkedList()


    def deleteAtEnd(self):
        if self.head == None:
            print("Head Node is not initialized!!")
            return
        
        if self.head.next == None:
            self.deleteAtBeginning()
            return

        temp = self.head

        while temp.next.next != None:
            temp = temp.next
        
        temp.next = None
        
        self.displayLinkedList()


    def deleteAtIndex(self, index):
        if self.head == None:
            print("Head Node is not initialized!!")
            return
        if index == 0:
            self.deleteAtBeginning()
            return
        pos = 0
        temp = self.head
        while temp != None and pos+1 != index:
            temp = temp.next
            pos += 1
        if temp != None and temp.next != None:
            temp.next = temp.next.next
        else:
            print(f"Index {index} is not valid!")
            return
        
        self.displayLinkedList()


    def deleteAtElement(self, element):
        deleted = 0
        if self.head == None:
            print("Head Node is not initialized!!")
            return
        temp = self.head
        
        if temp.data == element:
            deleted = 1
            self.deleteAtBeginning()
            return
        
        while temp.next != None:
            if temp.next.data == element:
                deleted = 1
                temp.next = temp.next.next
            else:
                temp = temp.next
        if deleted == 0:
            print(f"Element \'{element}\' is not presented in the list!")
            return
        
        self.displayLinkedList()

    def searchElement(self, element):
        found = 0
        pos = 0
        if self.head == None:
            print("Head Node is not initialized!!")
            return
        temp = self.head
        while temp != None:
            if temp.data == element:
                found = 1
                break
            else:
                pos += 1
                temp = temp.next
        if found == 1:
            print(f"Element \'{element}\' found at index {pos}.")
            return
        else:
            print(f"Element \'{element}\' has not found!")
            return
        
    def updateAtIndex(self, element, index):
        if self.head == None:
            print("Head node is not initialized!")
            return
        temp = self.head
        pos = 0
        while temp != None and pos != index:
            temp = temp.next
            pos += 1
        if temp:
            temp.data = element
        else:
            print(f"Index {index} is not valid!")
            return
        
        self.displayLinkedList()

    
    def updateAtElement(self, old_element, new_element):
        if self.head == None:
            print('Head node is not initialized!')
            return
        temp = self.head
        while temp != None and temp.data != old_element:
            temp = temp.next
        if temp:
            temp.data = new_element
        else:
            print(f'Element \'{old_element}\' is not found!')
            return
        
        self.displayLinkedList()



    def displayLinkedList(self):
        if self.head == None:
            print("Value Deleted.. Head node is not initialized!")
            return
        temp = self.head
        while temp:
            if temp.next is not None:
                print(f"{temp.data} -> ", end="")
            else:
                print(temp.data)
            temp = temp.next
```

```python
llist = SinglyLinkedList()

llist.insertAtBeginning('a')
llist.insertAtBeginning('b')
llist.insertAtEnd('c')
llist.insertAtEnd('z')
llist.insertAtIndex('d',3)
llist.insertAtIndex('e',4)

llist.deleteAtBeginning()
llist.deleteAtEnd()
llist.deleteAtIndex(3)
llist.deleteAtIndex(3)
llist.deleteAtIndex(6)
llist.deleteAtElement('e')
llist.searchElement('c')
llist.searchElement('f')
llist.insertAfterElement('b','c')
llist.updateAtIndex('z',2)
llist.updateAtElement('c','y')
```

```python
# Doubly Linked List

class Node:
    def __init__(self, value) -> None:
        self.prev = None
        self.data = value
        self.next = None

class DoublyLinkedList:
    def __init__(self) -> None:
        self.head = None


    def insertAtBeginning(self, data):
        new_node = Node(data)
        if self.head == None:
            self.head = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node
        
        self.displayLinkedList()

    
    def insertAtEnd(self, data):
        if self.head == None:
            self.insertAtBeginning(data)
            return
        new_node = Node(data)
        temp = self.head
        
        while temp.next != None:
            temp = temp.next

        temp.next = new_node
        new_node.prev = temp

        self.displayLinkedList()


    def insertAtPosition(self, data, index):
        if index == 0:
            self.insertAtBeginning(data)
            return
        
        new_node = Node(data)
        pos = 0
        temp = self.head

        while temp != None and pos+1 != index:
            temp = temp.next
            pos += 1
        
        if temp:
            new_node.prev = temp
            new_node.next = temp.next
            temp.next = new_node
            if new_node.next != None:
                new_node.next.prev = new_node
        else:
            print(f"Index {index} is not valid!")
            return

        self.displayLinkedList()

    
    def insertAfterElement(self, data, element):
        if self.head == None:
            print('Head node is not initialized!')
            return
        
        new_node = Node(data)
        temp = self.head

        while temp != None:
            if temp.data != element:
                temp = temp.next
            else:
                break

        if temp:
            new_node.prev = temp
            new_node.next = temp.next
            temp.next = new_node
            if new_node.next != None:
                new_node.next.prev = new_node
        else:
            print(f"Element \'{element}\' not found!")
            return
        
        self.displayLinkedList()

    
    def deleteAtBeginning(self):
        if self.head == None:
            print('Head node is not initialized!')
            return
        self.head = self.head.next

        self.displayLinkedList()


    def deleteAtEnd(self):
        if self.head == None:
            print('Head node is not initialized!')
            return
        
        if self.head.next == None:
            self.deleteAtBeginning()
            return
        
        temp = self.head
        while temp.next.next != None:
            temp = temp.next
        
        temp.next.prev = None
        temp.next = None

        self.displayLinkedList()


        
    def displayLinkedList(self):
        if self.head == None:
            print("Head node is not initialized!")
            return
        temp = self.head
        while temp:
            if temp.next is not None:
                print(f"{temp.data} -> ", end="")
            else:
                print(temp.data)
            temp = temp.next
```
