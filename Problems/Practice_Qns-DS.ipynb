1- Given an array of 1s and 0s which has all 1s first followed by all 0s. Find the number of 0s.

 Count the number of zeroes in the given array. Please try to solve it in O(log(n)).

```python
def count_zeroes(arr):
    left = 0
    right = len(arr) - 1
    
    while left <= right:
        mid = (left + right) // 2
        
        if arr[mid] == 1:
            left = mid + 1
        else:
            if mid == 0 or arr[mid - 1] == 1:
                return len(arr) - mid
            else:
                right = mid - 1
    
    return 0

arr = [1, 0, 0, 0, 0, 0, 0, 0]
zeroes_count = count_zeroes(arr)
print(zeroes_count)
```

2-  Given an integer X, find its square root. If X is not a perfect square, then return floor(âˆšx). Do it in O(log(n)) time.

```python
def find_square_root(X):
    if X == 0 or X == 1:
        return X
    
    left = 1
    right = X
    
    while left <= right:
        mid = (left + right) // 2
        
        if mid * mid == X:
            return mid
        elif mid * mid < X:
            left = mid + 1
            result = mid
        else:
            right = mid - 1
    
    return result

X = 16
square_root = find_square_root(X)
print(square_root)
```

3-  Reverse a Queue using Recursion.

```python
def rev_queue(arr):
    #reverse a queue using recursion
    if len(arr) == 0:
        return arr
    else:
        front = arr[0]
        arr.pop(0)
        arr = rev_queue(arr)
        arr.append(front)
        return arr

arr = [1,2,3,4,5]
print(rev_queue(arr))
```

 4- You are given two strings s and t.


String t is generated by random shuffling string s and then add one more letter at a random position.



Return the letter that was added to t

```python
def random_shuffle(s,t):
    s = sorted(s)
    t = sorted(t)
    i = 0
    while len(t) != 0:
        element = t[i]
        if element in s:
            s.remove(element)
            t.remove(element)
        else:
            return element

s = "asdfg"
t = "hgfdsa"
print(random_shuffle(s,t))
```
